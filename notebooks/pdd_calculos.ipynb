```python
import pandas as pd
import os
from zipfile import ZipFile 
user_profile = os.environ['USERPROFILE']
```

```python
path_zip = user_profile + r"\Downloads\moov.zip"

dfs = []
with ZipFile(path_zip) as z:
    # Lista todos os arquivos no ZIP
    for filename in z.namelist():
        # Verifica se Ã© um arquivo CSV
        if filename.endswith('.csv'):
            with z.open(filename) as f:
                df = pd.read_csv(f, 
                                encoding='ISO-8859-1',
                                on_bad_lines='skip', 
                                delimiter=';',
                                decimal=',', 
                                thousands='.',  
                                low_memory=False)
                dfs.append(df)
df_estoque = pd.concat(dfs, ignore_index=True)
```

```python
df_pdd = df_estoque[['VALOR_PRESENTE','VALOR_PDD','SITUACAO_RECEBIVEL','FAIXA_PDD']]
```

```python
atraso_vp = df_estoque[df_estoque['SITUACAO_RECEBIVEL'] == 'Vencido']['VALOR_PRESENTE'].sum()
pdd_sem_wop = df_estoque[df_estoque['FAIXA_PDD'] != 'WOP']['VALOR_PDD'].sum()
pdd_wop = df_estoque[df_estoque['FAIXA_PDD'] == 'WOP']['VALOR_PDD'].sum()
valor_presente = df_estoque['VALOR_PRESENTE'].sum()
total_pdd = df_estoque['VALOR_PDD'].sum()
```

```python
df_pdd = pd.DataFrame({
    'Carteira Atraso': [atraso_vp],
    'PDD s/ WO': [pdd_sem_wop],
    'PDD / Carteira Atraso': [pdd_sem_wop / atraso_vp if atraso_vp != 0 else 0],
    'Write-Off (WO)': [pdd_wop],
    'Carteira VP': [valor_presente],
    'PDD Total': [total_pdd],
    'PDD / Carteira': [total_pdd / valor_presente if valor_presente != 0 else 0]
})
```
