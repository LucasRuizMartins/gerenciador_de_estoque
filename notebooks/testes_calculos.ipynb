```python
import pandas as pd
import os
from zipfile import ZipFile 
```

```python
user_profile = os.environ["USERPROFILE"]
path_base = os.path.join(user_profile, 'Carmel Capital', 'Arquivos - Documentos', 
                         '00 - CARMEL ASSET', '01 - OPERACIONAL', 'TECNOLOGIA', 'BASE_ESTOQUE_ANALISE')
arquivo = 'base_virtus.zip'

path = rf"{path_base}\\{arquivo}"
```

```python
dfs = []
with ZipFile(path) as z:
    # Lista todos os arquivos no ZIP
    for filename in z.namelist():
        # Verifica se Ã© um arquivo CSV
        if filename.endswith('.csv'):
            with z.open(filename) as f:
                df = pd.read_csv(f, 
                                encoding='ISO-8859-1',
                                on_bad_lines='skip', 
                                delimiter=';',
                                decimal=',', 
                                thousands='.',  
                                low_memory=False)
                dfs.append(df)


df_estoque = pd.concat(dfs, ignore_index=True)
```
